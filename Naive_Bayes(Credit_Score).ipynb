```python
# to fill the objects we use mode
a=df['Gender'].mode()[0]
df['Gender'].fillna(a,inplace=True)

b=df['Income'].mean()
df['Income'].fillna(b,inplace=True)
df.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
lab=LabelEncoder()
df['Gender']=lab.fit_transform(df['Gender'])
df['Gender']
```

```python
df['Education']=lab.fit_transform(df['Education'])
df['Education']
```

```python
df['Marital Status']=lab.fit_transform(df['Marital Status'])
df['Marital Status']
```

```python
df['Home Ownership']=lab.fit_transform(df['Home Ownership'])
df['Home Ownership']
```

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
```

```python
# Model Creation
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
score=accuracy_score(y_test,y_pred)
score
```
