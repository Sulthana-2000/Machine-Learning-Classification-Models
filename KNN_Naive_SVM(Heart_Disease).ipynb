```python
mean_cp=df['cp'].mean()
df['cp']=df['cp'].fillna(mean_cp)

mean_fbs=df['fbs'].mean()
df['fbs']=df['fbs'].fillna(mean_fbs)

mean_restecg=df['restecg'].mean()
df['restecg']=df['restecg'].fillna(mean_restecg)

mean_oldpeak=df['oldpeak'].mean()
df['oldpeak']=df['oldpeak'].fillna(mean_oldpeak)
```

```python
df.isna().sum()
```

```python
x=df.iloc[:,:-1].values
x
```

```python
y=df.iloc[:,-1].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
y_train
```

```python
x_test
```

```python
# Normalization
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
x_train
```

```python
# KNN Classification
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=7)
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
```

```python
# Naive Bayes Classification
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
# SVM classification
from sklearn.svm import SVC
model=SVC()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
# these three models can be created in one file using "FOR" loop

from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix,accuracy_score

k_model=KNeighborsClassifier(n_neighbors=7)
n_model=GaussianNB()
s_model=SVC()
lst_model=[k_model,n_model,s_model]
```

```python
for i in lst_model:
  print("\nModel Name is :",i)
  i.fit(x_train,y_train)
  y_pred=i.predict(x_test)
  print("**********************")
  print(confusion_matrix(y_test,y_pred))
  print("Accuracy Score.........")
  print(accuracy_score(y_test,y_pred))

```
