```python
outlook=['Sunny','Sunny','Overcast','Rain','Rain','Rain','Overcast','Sunny','Sunny','Rain','Sunny','Overcast','Overcast','Rain']
temperature=['hot','hot','hot','mild','cool','cool','cool','mild','cool','mild','mild','mild','hot','mild']
humidity=['high','high','high','high','normal','normal','normal','high','normal','normal','normal','high','normal','high']
wind=['weak','strong','weak','weak','weak','strong','strong','weak','weak','weak','strong','strong','weak','strong']

play=['no','no','yes','yes','yes','no','yes','no','yes','yes','yes','yes','yes','no']
```

```python
# Here the given list are all objects so first of all, convert all object data into numerical data while creating a machine learning model
# Encoding (technique)
# 1.Label encoding
# 2.one hot encoding
# 3.Get dummies
```

```python
# 1.label encoding
from sklearn.preprocessing import LabelEncoder
lab=LabelEncoder()
outlook_encoder=lab.fit_transform(outlook)
outlook_encoder
```

```python
temp_encoder=lab.fit_transform(temperature)
temp_encoder
```

```python
humidity_encoder=lab.fit_transform(humidity)
humidity_encoder
```

```python
wind_encoder=lab.fit_transform(wind)
wind_encoder
```

```python
#to combine whole the list into one we use ZIP fn
feature=list(zip(outlook_encoder,temp_encoder,humidity_encoder,wind_encoder))
feature
```

```python
# model creation==== Naive Bayes algorithm
# 3 types of Classifier
# 1.Gaussian : continuous data(no strings/ objects)
# 2.Multinomial : discrete data
# 3.Bernoulli : combination of both cts and discrete
```

```python
# here we use MULTINOMIAL classifier
from sklearn.naive_bayes import MultinomialNB
data=MultinomialNB()
data.fit(feature,play)
print(data.predict([[2,0,0,0]]))
```
