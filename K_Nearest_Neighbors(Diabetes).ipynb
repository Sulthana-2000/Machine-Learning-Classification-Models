```python
# import the data and libraries
import numpy as np
import pandas as pd
df=pd.read_csv('/content/diabetes.csv')
df
```

```python
#to find missing values
df.isna().sum()
```

```python
x=df.iloc[:,:-1] # x should be always 2D array
x
```

```python
# to convert x as an array
x=df.iloc[:,:-1].values
x
```

```python
y=df.iloc[:,-1] # y should be always 1D array
y
```

```python
# to convert y as an array
y=df.iloc[:,-1].values
```

```python
# Split the data (Training and testing data conversion)
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
# Normalization Step (converting all values into same range)
# standard scalar(method/technique used)
# z=(x-u)/s
# x=input data
# u=mean of training data
# s=std deviation of training data

from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
x_train
```

```python
x_test
```

```python
# Model Creation
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=7)
knn.fit(x_train,y_train) #machine learning model (creation)
y_pred=knn.predict(x_test)
y_pred
```

```python
y_test
```

```python
# Performance evaluation ==== acuuracy score(confusion matrix)

from sklearn.metrics import confusion_matrix
mat=confusion_matrix(y_test,y_pred)
mat
```

```python
# Accuracy Score
from sklearn.metrics import accuracy_score
score=accuracy_score(y_test,y_pred)
score

#70% accuracy indicates that the ML model is good to use
```

```python
# here we are predicting using the model by giving the input data
pred=knn.predict(scaler.transform([[3,130,75,35,3,26,0.16,35]]))
pred
```

```python
# All Evaluation measures in one table
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
